# Step 6 - Find your Data

```python
# title
# subtitle
# author
# publisher
# publication date
# price
```

```python
len(response.json()['books'])
```

```python
results_json = response.json()['books']
```

```python
results_json[0]
```

```python
# title
for i in range(len(results_json)):
    # title
    title = results_json[i]['title']
    print(title)
    
    # subtitle
    subtitle = results_json[i]['subtitle']
    
    # authors
    authors = results_json[i]['authors']
    for author in authors:
        author_name = author['author_name']
        print(author_name)
        
    # publisher
    publisher = results_json[i]['publisher']
    print(publisher)
    
    # publication year
    publication_year = results_json[i]['publication_year']
    print(publication_year)
    
    # price
    price = results_json[i]['price']
    print(price)
    print("\n\n")
    
    # print(authors)
```

# Step 7 - Put everything together - Loop through results and append data inside a list

```python
title = []
subtitle = []
author = []
publisher = []
publication_year = []
price = []

for result in results_json:
    
    # title
    title.append(result['title'])
    
    # subtitle
    subtitle.append(result['subtitle'])
    
    # author
    author.append(result['authors'][0]['author_name'])
    
    # publisher
    publisher.append(result['publisher'])
    
    
    # publication year
    publication_year.append(result['publication_year'])
    
    # price
    price.append(result['price'])
```

# Step 8 - Create Pandas DataFrame

```python

books_dict = {
    'Title': title,
    'Subtitle': subtitle,
    'Author': author,
    'Publisher': publisher,
    'Publication Year': publication_year,
    'Price': price
}

books_df = pd.DataFrame(books_dict)
```
